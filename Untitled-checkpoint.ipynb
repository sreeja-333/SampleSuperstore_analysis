## Data Visualisation

```python
plt.title('Category')
plt.pie(data['Category'].value_counts(), labels=data['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
plt.title('Ship Mode')
plt.pie(data['Ship Mode'].value_counts(), labels=data['Ship Mode'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
plt.title('Region')
plt.pie(data['Region'].value_counts(), labels=data['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
state_sales = data.groupby("State")["Sales"].sum()

# Sort the states based on the sum of sales in descending order and select the top 10
top_10_states = state_sales.sort_values(ascending=False).head(10)
last_10_states = state_sales.sort_values(ascending=False).tail(10)

# Plot the bar chart
fig, axes = plt.subplots(1, 2, figsize=(9, 5))
top_10_states.plot(kind='bar', ax=axes[0])
last_10_states.plot(kind='bar',ax=axes[1])
axes[0].set_title('Top 10 countries sales')
axes[1].set_title('Least 10 countries sales')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize= (10,19))
data.groupby('Category')[['Profit', 'Sales']].agg(['sum']).plot.bar()
plt.legend(['Profits','Sales'])

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# First subplot: Discount & Profit relation based on Sub-Category
data.groupby('Sub-Category')[['Discount', 'Profit']].sum().plot(kind='bar', ax=axes[0])
axes[0].set_title('Discount & Profit Relation based on Sub-Category')

# Second subplot: Quantity & Profit relation based on Sub-Category
data.groupby('Sub-Category')[['Quantity', 'Profit']].sum().plot(kind='bar', ax=axes[1])
axes[1].set_title('Quantity & Profit Relation based on Sub-Category')

# plt.xticks(rotation=90)
plt.show()
```

```python
fig.suptitle("Total Sales VS Quantity ")
data.groupby('Sub-Category')[['Sales','Quantity']].agg(['sum']).plot(kind='bar')
plt.xticks(rotation=90)
plt.legend(['Sales','Quantity'])
plt.show()
     
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Sales with different shipping modes and Segments", fontsize=16)
sns.lineplot(x='Ship Mode', y='Sales', data=data, ax=axes[0])
sns.lineplot(x='Segment', y='Sales', data=data, ax=axes[1])
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.lineplot(x=data['Discount'], y=data['Profit'], data=data)
```

```python
plt.style.use('seaborn')
data.plot(kind = "scatter",figsize = (7,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 11, colormap = "cividis_r")
plt.ylabel('Total Profits',fontsize = 11)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 11)

plt.show()
```
